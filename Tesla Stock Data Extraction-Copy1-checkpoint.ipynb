# Tesla Stock Data Extraction

```python
# 导入 yfinance 库
import yfinance as yf

# 选择特斯拉股票的代码和时间范围
symbol = "TSLA"
start = "2023-01-01"
end = "2024-01-31"

# 使用 download 方法获取股票数据
data = yf.download(symbol, start, end)

# 打印数据的前五行
data.head()
```

```python
data.tail()
```

```python
# 导入 requests 和 BeautifulSoup 库
import requests
from bs4 import BeautifulSoup
# 定义目标网站的 URL
url = "https://ir.tesla.com/financial-information/quarterly-results"
# 发送 HTTP 请求，获取目标网站的响应内容
response = requests.get(url)
# 判断响应状态码是否为 200，表示请求成功
if response.status_code == 200:
    # 解析响应内容，创建 BeautifulSoup 对象
    soup = BeautifulSoup(response.text, "html.parser")
    # 使用 CSS 选择器，找到包含特斯拉收入数据的表格
    table = soup.select_one("table.wpb_table")
    # 使用 find_all 方法，找到表格中的所有行
    rows = table.find_all("tr")
    # 创建一个空列表，用于存储特斯拉收入数据
    tesla_revenue = []
    # 遍历每一行，提取特斯拉收入数据
    for row in rows:
        # 使用 find_all 方法，找到行中的所有单元格
        cells = row.find_all("td")
        # 创建一个空列表，用于存储一行的数据
        row_data = []
        # 遍历每一个单元格，提取其文本内容
        for cell in cells:
            row_data.append(cell.text.strip())
        # 将一行的数据添加到特斯拉收入数据列表中
        tesla_revenue.append(row_data)
    # 打印特斯拉收入数据列表
    print(tesla_revenue)
else:
    # 打印请求失败的提示信息
    print("请求失败，状态码为", response.status_code)
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import yfinance as yf

def make_graph(stock_data, title):

    plt.figure(figsize=(10, 6))  # 设置图表大小
    plt.plot(stock_data['Close'],color="black")  # 绘制收盘价曲线
    plt.title(title)  # 设置图表标题
    plt.xlabel('Date')  # 设置x轴标签
    plt.ylabel('Close Price (USD)')  # 设置y轴标签
    plt.show()

# 使用yfinance获取特斯拉股票数据
ticker_symbol = 'TSLA'
ticker_data = yf.Ticker(ticker_symbol)
tesla_data = ticker_data.history(period="1y")  # 获取过去一年的数据

# 使用make_graph函数绘制特斯拉股票数据图表
make_graph(tesla_data, "Tesla Stock Data - Past Year")
```

```python
import requests
from bs4 import BeautifulSoup

# 定义目标网站的 URL
url = "https://www.199it.com/archives/1415471.html"

# 发送 HTTP 请求，获取目标网站的响应内容
response = requests.get(url)

# 判断响应状态码是否为 200，表示请求成功
if response.status_code == 200:
    # 解析响应内容，创建 BeautifulSoup 对象
    soup = BeautifulSoup(response.text, "html.parser")

    # 尝试使用更具体或者不同的CSS选择器
    # 这里只是一个示例，根据实际页面结构进行调整
    table = soup.select_one("article table")

    if table:
        # 使用 find_all 方法，找到表格中的所有行
        rows = table.find_all("tr")

        # 遍历每一行，提取数据
        for row in rows:
            # 使用 find_all 方法，找到行中的所有单元格
            cells = row.find_all("td")

            # 遍历每一个单元格，打印其文本内容
            for cell in cells:
                print(cell.text, end="\t")

            # 换行
            print()
    else:
        print("未找到表格")
else:
    # 打印请求失败的提示信息
    print("请求失败，状态码为", response.status_code)
```
